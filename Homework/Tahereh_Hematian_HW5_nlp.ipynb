- Build 'word frequency'. Make sure you have removed the stopwords
- Determine the maximum frequency as 'word_frequencies[word]=
(word_frequencies[word]/maximum_frequency).

```python
# Building work frequency
Freq_word = {}
for phrase in docx:
  all_words = [word.text for word in phrase]
  for w in all_words:
    w1 = w.lower()
    if w1 not in extra_words and w1.isalpha():
      if w1 in Freq_word.keys():
        Freq_word[w1]+=2
      else:
        Freq_word[w1] = 1
Freq_word
```

```python
# Max 2 ferquency
val = sorted(Freq_word.values())[::-1]
max_freq = val[:2]
max_freq
```

```python
# what are these 2 max frequenies
for word,freq in Freq_word.items():
  if freq in max_freq:
    print(word ,end=" ")
  else:
    continue
```

```python
# Max 3 frequencies
val=sorted(Freq_word.values())
max_freq=val[-3:]
print("Topic of document given :")
for word,freq in Freq_word.items():
  if freq in max_freq:
    print(word ,end=" ")
  else:
    continue
```

- Determine the maximum frequency as 'word_frequencies[word]=
(word_frequencies[word]/maximum_frequency).

```python
# Normalize max frequency
Freq_word_norm = {}
for word in Freq_word.keys():
       Freq_word_norm[word] = (Freq_word[word]/max_freq[-1])
Freq_word_norm
```

- Tokenize the sentences. Generate the sentence_scores. Score every sentence
based on number of words.


```python
# frequency of each words in a sentence
sent_strength={}
for phrase in docx:
  for sent in phrase.sents:
    for word in sent :
      if word.text.lower() in Freq_word.keys():
        if sent in sent_strength.keys():
          sent_strength[sent]+=Freq_word[word.text.lower()]
        else:
          sent_strength[sent]=Freq_word[word.text.lower()]
      else: 
        continue  
sent_strength
```

- Import nlargest from heapq and provide 'summarized_sentences'.

```python
import heapq
```

```python
heap = [(-value, key) for key,value in sent_strength.items()]
largest = heapq.nsmallest(10, heap)
largest = [(key, -value) for value, key in largest]
```

```python
largest
```

```python
top_sentences=(sorted(sent_strength.values())[::-1])
top30percent_sentence=int(0.3*len(top_sentences))
top_sent=top_sentences[:top30percent_sentence]
```

```python
# Summarized sentences
summary=[]
for sent,strength in sent_strength.items():
       if strength in top_sent:
          summary.append(sent)
       else:
          continue
for i in summary:
  print(i,end="")
```
